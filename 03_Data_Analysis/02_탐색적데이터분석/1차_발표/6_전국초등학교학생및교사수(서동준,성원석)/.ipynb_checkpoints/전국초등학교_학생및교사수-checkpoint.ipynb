## 0. 초기 가설 설정

#### 데이터를 확인하기 이전 기존에 알고있던 지식을 바탕으로 초기가설을 세워보았습니다.

1. 저출산이 가속화 되고 있기 때문에 학생수가 크게 줄었을 것이다.
2. 지방 학교 / 학생수가 줄어들고 수도권 및 광역시 학교 / 학생수가 늘었을 것이다.
3. 전체적으로는 학교 수가 줄어들것으로 예상되고, 특히 국공립 학교수가 줄어들었을 것이다.
4. 지방 교대 입결 수치가 낮아지고 입시 결과 합격인원 미달이 발생하는 학교가 늘어나는 것을 보아 지방 교사 인원수가 줄어들거나 미미하게 증가하고, 수도권 및 광역시 교사 인원수가 더 많이 증가할 것으로 추정된다.
5. 학생 수가 많은 곳의 학군이 잘 알려진 학구열 높은 메이저 학군일 것이다. (강남, 수성구)
6. 6번과 마찬 학교 수가 많은 곳의 학군이 잘 알려진 학구열 높은 메이저 학군일 것이다. (강남, 수성구)
7. 특수학교 / 순회학교 학급수가 특수교사 부족현상으로 인해 줄어들었을 것이다.

#### 가설검증 1)

```python
# 2014년 / 2024년 전체 학생 수 비교
```

```python
students = pd.DataFrame(ps.groupby('연도')['학생수(계)'].sum())
sns.barplot(data = students, y = '연도', x= '학생수(계)')
plt.show()
```

```python
# 학생수가 10% 정도 감소한 것을 확인할 수 있다.
# 2014년 기준 1학년이 2007년생, 2024년 기준 1학년이 2017년생인 것을 감안,
# 출산율이 본격적으로 감소추세에 접어든 것이 2015년이기 때문에
# 생각보다 학생수 하락폭이 많이 크지 않은 것으로 확인되었다.
```

---

```python
# 2014 / 2024년 학년별 학생수 비교
```

```python
ps_sum = ps.groupby('연도').sum()
ps_sum.reset_index(inplace = True)
```

```python
ps_sum_melted = ps_sum.melt(id_vars=['연도'], value_vars=['6학년 학생수', '5학년 학생수', '4학년 학생수', '3학년 학생수', '2학년 학생수', '1학년 학생수'],
                    var_name='학년', value_name='학생수')
plt.figure(figsize=(10, 6))
sns.barplot(x = '연도', y = '학생수', hue = '학년', data = ps_sum_melted, palette = 'Set1')
plt.xlabel('연도')
plt.ylabel('학생수')
plt.title('연도별 학년별 학생수')
plt.legend(title='학년')
plt.show()
```

```python
# 2014년 데이터를 보면 뚜렷한 증가/감소 추세가 보이지 않으나
# 2024년 데이터를 보면 점차 감소하는 추세가 나타나 보인다.
```

```python
# 2014년 학생수 // 학급수
sc_ratio_2014 = students.iloc[0][0] // classes.iloc[0][0]
# 2024년 학생수 // 학급수
sc_ratio_2024 = students.iloc[1][0] // classes.iloc[1][0]

print(f'2014년 학급당 학생수: {sc_ratio_2014}\n2024년 학급당 학생수: {sc_ratio_2024}')
```

```python
# 학생수가 줄고 학급수가 늘면서 학급당 학생수가 줄어든 것을 확인할 수 있다.
```

```python
public_private = ps.groupby('연도')['설립구분'].value_counts()
fig, axes = plt.subplots(1, len(public_private.index.levels[0]), figsize = (12, 6))
for i, (year, group) in enumerate(public_private.groupby(level = 0)):
    ax = axes[i]
    group_data = group.droplevel(0)
    ax.pie(group_data, labels = group_data.index, autopct = '%.1f%%')
    ax.set_title(f'{year}년')
plt.show()
```

```python
# 전체 학교 중 공립학교 비율이 압도적으로 높은 것을 확인할 수 있고,
# 지방인구 유출과 관련해 지방 국공립 학교의 통폐합 등으로 국공립 학교 수가 줄어들 것으로 예상했으나
# 오히려 공립학교수가 늘어나고 사립학교 수가 줄어든 것을 확인할 수 있다.
```

## 4. 탐색적연구

#### 가설에 대한 검증 결과를 토대로 결론을 내려보았습니다.

1. 2014년 학생 기준 2002-2007년생, 2024년 학생 기준 2012-2017년생 으로 구성되어있기 때문에 출산율이 감소추세에 들기 시작한 2015년을 전후로 학생수가 본격적으로 줄어들기 시작하는 것으로 추정되기 때문에 3학년 학생수부터 본격적으로 줄어들기 시작하는것으로 보인다.
2. 학교 수는 오히려 늘어난 것을 확인할 수 있었다. 국공립 학교 통폐합을 예상했으나 오히려 공립학교 수가 늘어났고 사립학교 수가 줄어든 것을 확인할 수 있었다.
3. 당초 지방 학생수가 줄고 수도권 및 광역시 학생수가 늘것이라 예상한것과 달리 제주특별자치도, 세종특별자치시를 제외한 모든 지역에서 학생수가 감소한것을 확인할수 있었다. 특히 세종특별자치시 학생수가 4배로 뛰어 디테일한 조사가 필요하다.
4. 교사수는 증가하였으나 서울, 대전, 대구교육청 소속 학교의 교사수는 오히려 감소한것을 확인할 수 있었다. 그리고 세종시가 기존 교사수 대비 300% 가까이 증가한 것을 확인할 수 있었다.
5. 특수교사 부족 여부와 상관없이 특수학급 학생수가 증가함에 따라 학급수도 증가한 모습을 확인할 수 있었다. 또한 병원이나 가정에 특수교사가 방문하여 진행하는 순회학급의 경우 학생수 및 학급수가 줄어든 것을 확인할 수 있었다.
6. 유명한 학구열 메이저 도시인 강남쪽이 학교가 많을줄 알았으나 예상외로 경기도 지역에 학교가 많은것으로 판단되었다. 경기도에 많은 인구수가 있다보니 학교수가 많은것으로 판단되었다.
