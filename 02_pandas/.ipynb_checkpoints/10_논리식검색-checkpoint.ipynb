# 논리식 검색을 통한 데이터 필터링

* 필터링 조건: '도수(명)' 열에 20000 보다 작은 행을 필터링 할 것

```python
#  사전 작업으로 각 행의 논리식 브로드캐스트 연산
(sat['도수(명)']<20000).head()
```

* 논리식 검색 조건을 행의 필터링 조건으로 활용

```python
sat[sat['도수(명)']<20000]
```

```python
* 열에 
```

```python
sat[sat['비율(%)'] > 10.0]
```

* 다중 논리식(조건식) 검색
    * ( 조건식1 & 조건식2 .... 조건식N )  
 A & B & C
    * ( 조건식1 (& 또는 |) 조건식2 ... 조건식 N)  
A & B | C

```python
[주의] 개별 조건은 항상 ()로 안전하게 감싸자.
```

* 문제 한 번 풀어 봅시다!  
'도수(명)' 열의 값이 20000 보다 작고 '비율(%)' 열의 값이 5.0 보다 작은 모든 행을 필터링 하세요.  
sat[sat['도수(명)'] < 20000 ]  <- 이것을 확장해 보세요

```python
sat[(sat['도수(명)'] < 20000) & (sat['비율(%)'] < 5.0)]
```

* 문제 한 번 풀어 봅시다!  
'도수(명)' 열의 값이 20000 보다 작거나 '비율(%)' 열의 값이 10.0 보다 작은 모든 행을 필터링 하세요.  
sat[sat['도수(명)'] < 20000 ]  <- 이것을 확장해 보세요

```python
sat[(sat['도수(명)'] < 20000) | (sat['비율(%)'] < 10.0)]
```
