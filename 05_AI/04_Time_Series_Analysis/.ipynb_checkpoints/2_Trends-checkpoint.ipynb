# 주요 구성 요소

```python
import pandas as pd
from neuralprophet import NeuralProphet, set_log_level

# Disable logging messages unless there is an error
set_log_level("ERROR")

# Load the dataset from the CSV file using pandas
df = pd.read_csv("https://github.com/ourownstory/neuralprophet-data/raw/main/kaggle-energy/datasets/tutorial01.csv")
```

* NeuralProphet는 기본적으로 Trend와 Seasonality 구성 요소가 활성화되어 있다.
* 이 두 번째 튜토리얼에서는 Trend 구성 요소를 더 자세히 살펴보고 가장 기본적인 설정으로 시작한다.
* 따라서 현재는 Seasonality 구성 요소를 비활성화하고 Trend의 변화점 개수를 0으로 설정한다."

# 트랜드

* 추세는 데이터가 오래 기간 동안 어떻게 증가하거나 감소하는지를 나타냅니다. 
* 추세는 데이터의 장기적인 변화를 보여주며, 주로 시간에 따른 데이터의 일반적인 경향성을 표현합니다
* 추세는 추세가 변하는 변화점(change point)가 존재하며 아래 n_changepoints 인자로 강제 조정할 수 있다.

```python
# 모델과 예측
m = NeuralProphet(
    # 변화점 변경 비활성화
    n_changepoints=0,
    # 계절성 구성요소 비활성화
    yearly_seasonality=False,
    weekly_seasonality=False,
    daily_seasonality=False,
)
```

```python
# plotly 라이브러리를 통한 시각화
m.set_plotting_backend("plotly")
```

```python
metrics = m.fit(df)
forecast = m.predict(df)
```

```python
m.plot(forecast)
```

```python
# Model and prediction
m = NeuralProphet(
    # Use default number of change trendpoints (10)
    # n_changepoints=0,
    # Disable seasonality components
    yearly_seasonality=False,
    weekly_seasonality=False,
    daily_seasonality=False,
)
# m.set_plotting_backend("matplotlib")  # Use matplotlib due to #1235
m.set_plotting_backend("plotly")
metrics = m.fit(df)
forecast = m.predict(df)
```

```python
m.plot(forecast)
```

* 아래 트랜드 분석 결과 9개의 변곡점이 있는 것으로 파악된다.

```python
m.plot_parameters(components=["trend"])
```
