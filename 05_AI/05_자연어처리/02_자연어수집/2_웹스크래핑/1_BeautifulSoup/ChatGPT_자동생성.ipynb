```python
import pandas as pd
from bs4 import BeautifulSoup

# HTML 파일 열기
with open("프로야구순위_24-08-22.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# BeautifulSoup으로 HTML 파싱
soup = BeautifulSoup(html_content, 'html.parser')

# 테이블 찾기: class가 'tb_type2'인 테이블을 선택
table = soup.find("table", class_="tb_type2")

# 테이블의 행 데이터 추출
rows = []
for row in table.find_all("tr"):
    columns = row.find_all(["th", "td"])  # th(헤더)와 td(데이터)를 모두 포함
    if columns:  # 빈 행 제외
        rows.append([col.get_text(strip=True) for col in columns])

# pandas DataFrame으로 변환 (헤더가 따로 없는 경우 자동으로 열 이름을 부여)
df = pd.DataFrame(rows)

# DataFrame 출력
df
```

```python
import pandas as pd
from bs4 import BeautifulSoup

# HTML 파일 읽기
with open('프로야구순위_24-08-22.html', 'r', encoding='utf-8') as file:
    html_content = file.read()

# BeautifulSoup 객체 생성
soup = BeautifulSoup(html_content, 'html.parser')

# 특정 테이블 추출 (id가 teamRankTabPanel_0인 부분에서 첫 번째 테이블을 찾습니다)
table = soup.find('div', {'id': 'teamRankTabPanel_0'}).find('table')

# 테이블에서 데이터를 추출하여 데이터프레임으로 변환
df = pd.read_html(str(table))[0]

# 데이터프레임 출력
df
```
