# 신경망 모델 훈련

```python
# 실행마다 동일한 결과를 얻기 위해 케라스에 랜덤 시드를 사용하고 텐서플로 연산을 결정적으로 만듭니다. 
import tensorflow as tf

tf.keras.utils.set_random_seed(42)
tf.config.experimental.enable_op_determinism()
```

## 손실 곡선

```python
from tensorflow import keras
from sklearn.model_selection import train_test_split

(train_input, train_target), (test_input, test_target) = \
    keras.datasets.fashion_mnist.load_data()

train_scaled = train_input / 255.0

train_scaled, val_scaled, train_target, val_target = train_test_split(
    train_scaled, train_target, test_size=0.2, random_state=42)
```

```python
train_scaled.shape, val_scaled.shape, train_target.shape, val_target.shape
```

* 개별적으로 레이어를 추가하는 함수 작성

```python
def model_fn(a_layer=None):
    model = keras.Sequential()
    model.add(keras.layers.Flatten(input_shape=(28, 28)))
    model.add(keras.layers.Dense(100, activation='relu'))
    if a_layer:
        model.add(a_layer)
    model.add(keras.layers.Dense(10, activation='softmax'))
    return model
```

```python
model = model_fn()

model.summary()
# 아래 warning 신경망을 개별 추가하고 이후에 나오는 것으로 무시한다.
# 사용자 입력에 따라 특정 층을 추가하기 위해 의도적으로 만든 것이기 때문이다.
```

### 5 에포크(Epoch) 성능 확인

```python
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])

history = model.fit(train_scaled, train_target, epochs=5, verbose=0)
```

```python
print(history.history.keys())
```

```python
import matplotlib.pyplot as plt

plt.plot(history.history['loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

```python
plt.plot(history.history['accuracy'])
plt.xlabel('epoch')
plt.ylabel('accuracy')
plt.show()
```

## 검증 손실

```python
model = model_fn()
model.compile(loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# 훈련데이터와 시험데이터셋의 학습및 평가결과를 동시에 비교하기 위해서
# validation_data옵션에 시험데이터 정보를 입력
# => 모델 학습은 train_target으로 하지만 검증데이터의 결과도 지정하여 확인할 수 있다.
history = model.fit(train_scaled, train_target, epochs=20, verbose=0, 
                    validation_data=(val_scaled, val_target))
```

## 콜백
* 훈련 과정 중간에 특정 작업을 수행할 수 있게 하는 객체
* 주로 조기 종료의 용도로 활용한다.

* 현재 버전에서 콜벡 기능에서 ModelCheckpoint함수가 h5, keras 확장자를 안정적으로 지원하지 않고 있다.  
 그래서 아래 코드는 향후 안정화 되면 동작 가능한 코드이지만 현재는 동작하지 않는다.
* ModelCheckpoint: 학습도중에 가장 성능이 좋은 모델을 저장하는 기능을 수행 

```python
# model = model_fn(keras.layers.Dropout(0.3))
# model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', 
#               metrics='accuracy')

# # 훈련 수행하는 동안 손실값이 가장 적은 모델을 저장
# checkpoint_cb = keras.callbacks.ModelCheckpoint('best-model.h5',
# # checkpoint_cb = keras.callbacks.ModelCheckpoint('best-model.keras', 
#                                                 save_best_only=True)

# model.fit(train_scaled, train_target, epochs=2, verbose=0, 
#           validation_data=(val_scaled, val_target),
#           callbacks=[checkpoint_cb])
```

* 아래 코드는 최신 포멧인 keras를 유지하면서 Model Checkpoint를 사용자 정의

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
def save_model_checkpoint(epoch, logs):
    if logs['val_loss'] < save_model_checkpoint.best_val_loss:
        save_model_checkpoint.best_val_loss = logs['val_loss']
        model.save('best-model.keras')
        print('Model checkpoint saved.')

save_model_checkpoint.best_val_loss = float('inf')

model.fit(train_scaled, train_target, epochs=20, verbose=0, 
          validation_data=(val_scaled, val_target),
    # Other model.fit arguments...
    callbacks=[tf.keras.callbacks.LambdaCallback(on_epoch_end=save_model_checkpoint)]
 )
```

```python
# model = keras.models.load_model('best-model.h5')
model = keras.models.load_model('best-model.keras')

model.evaluate(val_scaled, val_target)
```

### 조기종료

```python
# model = model_fn(keras.layers.Dropout(0.3))
# model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', 
#               metrics='accuracy')

# checkpoint_cb = keras.callbacks.ModelCheckpoint('best-model.h5', 
#                                                 save_best_only=True)
# # patience: 모델성능이 향상되지 않는(손실값이 줄어들지 않는) 것을 지속시킬 최대 에포크
# # restore_best_weights: EarlyStop시 최적의 가중치 조건으로 복원
# early_stopping_cb = keras.callbacks.EarlyStopping(patience=2,
#                                                   restore_best_weights=True)

# history = model.fit(train_scaled, train_target, epochs=2, verbose=0, 
#                     validation_data=(val_scaled, val_target),
#                     callbacks=[checkpoint_cb, early_stopping_cb]
                    
#                    )
```

```python
model = model_fn(keras.layers.Dropout(0.3))
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])

checkpoint_cb = keras.callbacks.ModelCheckpoint('best-model.keras', 
                                                save_best_only=True)
# patience: 모델성능이 향상되지 않는(손실값이 줄어들지 않는) 것을 지속시킬 최대 에포크
# restore_best_weights: EarlyStop시 최적의 가중치 조건으로 복원
early_stopping_cb = keras.callbacks.EarlyStopping(patience=2,
                                                  restore_best_weights=True)

history = model.fit(train_scaled, train_target, epochs=20, verbose=0, 
                    validation_data=(val_scaled, val_target),
#                     callbacks=[checkpoint_cb, early_stopping_cb]
                    callbacks=[tf.keras.callbacks.LambdaCallback(on_epoch_end=save_model_checkpoint), early_stopping_cb]
                   )
```

```python
print(early_stopping_cb.stopped_epoch)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.legend(['train', 'val'])
plt.show()
```

```python
model.evaluate(val_scaled, val_target)
```
