# 명시적 인덱서 검색

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head()
```

# loc

```python
데이터프레임을 생성시 부여된 명시적인(explicit) 행, 열 인덱스 정보를 활용해 데이터를 추출하는 방식
```

* 데이터프레임객체.loc[ 행검색조건, 열검색조건 ]

```python
diamonds.loc[0] # 첫번째 행 데이터, 행인덱스 0의 행데이터
```

```python
diamonds.loc[3] # 행인덱스 3의 행데이터
```

```python
diamonds.head(3)
```

```python
# 데이터프레임객체.loc[ 행검색조건, 열검색조건]
diamonds.loc[1,'price']
```

```python
diamonds.loc[2,'clarity']
```

* 슬라이싱 적용
* [주의] loc에서는 마지막 인덱스를 포함

```python
# 10행에서 25행 사이의 모든 'carat' 데이터 조회
# loc에서는 마지막 인덱스를 포함
diamonds.loc[ 10:25 , 'carat']
```

* loc에서 펜시검색 적용

```python
# 10행에서 25행 사이의 모든 'carat', 'depth' 열 데이터 조회
diamonds.loc[ 10:25 , ['carat', 'depth']]
```

# 암묵적(Implicity) 인덱서

* 인덱스 또는 colomns이 레이블(문자)로 존재하더라도 암묵적으로 인덱스 접근이 가능  
* iloc[행필터링조건,열필터링조건]  
* 암묵적 인덱스는 행 또는 열에 시작이 0부터 시작된다.  

```python
diamonds.head(1)
```

```python
diamonds.info()
```

```python
diamonds.shape
```

```python
# 중위수 데이터 조회
diamonds.iloc[26969]
```

```python
diamonds.columns
```

```python
# 모든 'price' 열 정보 추출
# temp = [1,2,3,4,5,6,7]
# temp[:] # 모든 데이터 가져오기
diamonds.iloc[ : , 6]
```

```python
# 비교
# diamonds['price'] : 간단한 열 추출은 이 방법으로
```

```python
# 중위수에서 10개의 다이아몬드 가격
# iloc에서느 마지막 인덱스 불포함
diamonds.iloc[ 26969:26979 , 6]
```

```python
diamonds.head()
```

```python
diamonds.tail()
```

* 마이너스 인덱스 활용

```python
diamonds.iloc[ 26969:26979 , -4]
```

* 마이너스 인덱스를 활용한 행, 열 슬라이싱  
마지막 3개열 중위수부터 10개

```python
diamonds.iloc[ 26969:26979 , -3:]
```

* 열이름을 통한 슬라이싱  
전제조건] 열이름을 통한 슬라이싱을 하려면 열이름이 정렬이 되어 있어야 한다.

```python
diamonds.head(1)
```

```python
diamonds.loc[:,'x':'z'].head()
```

```python
diamonds.loc[:,'carat':'color'].head()
```

# 인덱스 변경

```python
데이터 분석을 위해 기존의 열의 값을 인덱스로 활용
```

```python
diamonds.head()
```

```python
set_i = diamonds.set_index('color')
set_i
```

* 인덱스 정렬

```python
set_i.sort_index().head(20)
```

```python
set_i # sort_index() 실행하면 자기 자신의 값을 변경하지 않는다.
```

```python
sorted_set_i = set_i.sort_index()
sorted_set_i
```

```python
# 'E'~'F' 색상의 모든 다이아몬드 데이터 추출
sorted_set_i.loc['E':'F']
```

```python
set_i
```

```python
set_i.loc['E':'F'] # 인덱스가 정렬이 되지 않은 상황에서 슬라이싱 문법 적용시 에러 발생
```
