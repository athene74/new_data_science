```python
# 0. 사용할 패키지 불러오기
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Input

# 1.1 데이터셋 생성하기
x_train = np.random.random((1000, 1))
y_train = x_train * 2 + np.random.random((1000, 1)) / 3.0
x_test = np.random.random((100, 1))
y_test = x_test * 2 + np.random.random((100, 1)) / 3.0
```

```python
# 1.2 입력값 이해하기
import matplotlib.pyplot as plt

# 데이터셋 시각화하기
plt.scatter(x_train, y_train, alpha=0.5)
plt.title('Scatter plot of the training data')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
# 2. 모델 구성하기
model = Sequential()
model.add(Input(shape=(1,)))  # 'Input' 레이어로 입력 형태 정의
# model.add(Dense(64, input_dim=1, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(64, activation='relu'))
model.add(Dense(1))

# 3. 모델 학습과정 설정하기
model.compile(optimizer='adam', loss='mse')

# 4. 모델 학습시키기
hist = model.fit(x_train, y_train, epochs=50, batch_size=64)

# 5. 학습과정 살펴보기
import matplotlib.pyplot as plt

plt.plot(hist.history['loss'])
plt.ylim(0.0, 1.5)
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train'], loc='upper left')
plt.show()

# 6. 모델 평가하기
loss = model.evaluate(x_test, y_test, batch_size=32)
print('loss : ' + str(loss))
```

```python
# 7. 예측하기
input_value = np.array([[0.5]])  # 특정 특성 값을 포함하는 2D 배열
predicted_value = model.predict(input_value)
print(f'입력 값: {input_value.flatten()[0]}')
print(f'예측된 출력 값: {predicted_value.flatten()[0]}')
```
