### 넘파이의 random 함수

### 1. numpy.random.rand
 0~1 사이의 값의 uniform distribution 균일 분포, 균등 분포라고 불리는 값을 return 하는 함수

* 실생활에서 균등분포 데이터의 예
    * 주사위 던지기
    * 확률 추첨게임(로또)

```python
DATA_LIMIT = 1000000
```

```python
import pandas as pd
import numpy as np

np.random.rand(5)
```

```python
np.random.rand(2,5)
```

```python
rand_data = np.random.rand(DATA_LIMIT)
plt.hist(rand_data)
```

### 2. numpy.random.randn

 Normal(Gaussian) Distribution 정규 분포, 가우시안 분포라고 불리는 값을 return 하는 함수입니다. 
 평균 0, 분산이 1인 값을 반환

* 실생활에서 발생하는 정규분포 데이터예
    * 키, 몸무게, 성적, IQ

```python
randn_data = np.random.randn(DATA_LIMIT)
# bin: 히스토그램에서 표시되는 막대그래프의 갯수를 의미
# bin의 갯수가 크다면: 개별 구간이 좁아진다. 세밀해진다.
# bin의 갯수가 작다면, 막대그래프 갯수가 줄어들면서 개별 구간이 늘어진다.
# plt.hist(randn_data, bins=100) # 전체 데이터 구간을 100으로 설정
# plt.hist(randn_data, bins=200) # 전체 데이터 구간을 200으로 설정
plt.hist(randn_data, bins=50) # 전체 데이터 구간을 50으로 설정
```

# 히스토그램 응용 데이터 시뮬레이션

* 대한민국 20대 성인 남자 키 

```python
move_right =  174.5 # 통계나 가설에 의해서 정의한 값
variance_amplication = 7 # variance : 분산
```

```python
height_data = np.round(
    move_right + 
    (np.random.randn(DATA_LIMIT)*variance_amplication),
    1)
```

```python
height_data
```

```python
plt.xlabel('키')
plt.ylabel('누적개수')
plt.hist(height_data, bins=40)
```

```python
m_move_right =  74.3
w_move_right =  159.9
variance_amplication = 7
m_height_data = np.round(m_move_right + (np.random.randn(DATA_LIMIT)*variance_amplication),1)
w_height_data = np.round(w_move_right + (np.random.randn(DATA_LIMIT)*variance_amplication),1)

plt.xlabel('키')
plt.ylabel('누적개수')
plt.hist((m_height_data,w_height_data), bins=100, label=['성인 남자키','성인 여자키'])

plt.title('전국 성인남녀 키 현황 (2024.3 산업통산자원부 통계자료 시뮬레이션)')
plt.legend()

plt.show()
```

# 데이터 보정작업에서의 전략

```python
* 수집한 수치형 데이터를 시각화 해보니 정규분포형 데이터임을 알게 되었다.
* 전체 데이터는 100만건이고 이중에 약 1000여건의 결측치를 발견했다.
* 결측치의 갯수를 확인한다.
* 의사결정
    * 삭제(drop) : 결측치가 굉장히 많을 경우, 도메인 데이터 전문가 판단
    * 재조사 : 전체 데이터 대비 결측치가 상당히 많은 경우
    * 결측치를 보정하기 위한 단계
     * 해당 수치데이터의 평균을 알아야 한다.
     * 평균에 근거한 정규분포 데이트를 생성한다.(randn 함수를 사용하거나 평균값으로 대체한다.)
```

### * 대한민국 성인 남여 몸무게에 대한 데이터를 시뮬레이션하여 생성하고 히스토그램으로 변경해 보세요.

```python
m_move_right =  74.3
w_move_right =  58.6
variance_amplication = 6
m_height_data = np.round(m_move_right + (np.random.randn(DATA_LIMIT)*variance_amplication),1)
w_height_data = np.round(w_move_right + (np.random.randn(DATA_LIMIT)*variance_amplication),1)

plt.xlabel('몸무게')
plt.ylabel('누적개수')
plt.hist((m_height_data,w_height_data), bins=80, label=['성인 남자몸무게','성인 여자몸무게'])

plt.title('전국 성인남녀 몸무게 현황 (2024.7 대구알파시티 시뮬레이션)')
plt.legend()

plt.show()
```
