```python
movies_str = pd.read_csv('data/korea_movie_list.csv', encoding='cp949')
movies_str.head()
```

# 결측값 확인

* hasnans 속성: 해당 열에 NaN 값이 있다면 True, 없으면 False

```python
movies_str['movie_code'].hasnans
```

```python
movies_str['title_Eng'].hasnans
```

* isnull() : 개별 열에 대한 모든 결측값(NaN) 유무에 대해서 True/False를 반환

```python
movies_str['actor'].isnull().head(10)
```

* 결측값 열 확인

```python
movies_str.isnull().head(10)
```

```python
missing_values = movies_str.isnull().any()
missing_values
```

* 결측치 열만 필터링

```python
column_missing_values = missing_values [ missing_values == True]
column_missing_values
```

# 특정열의 값이 null인 전체 행 필터링

```python
movies_str[ 'actor' ].isnull().head(10)
```

```python
# 데이터프레임[블리언마스크필터링조건] : 블리언 마스크 필터링 조건: 전체 행별로 True/False 값으로 지정되며 True 인 값의 행이 필터링 됨
movies_str[movies_str[ 'actor' ].isnull()].head(20)
```

```python
# 'title_Eng'열의 값 중 정의 되지 않은 모든 행에 대해서 필터링 해보세요.
movies_str[movies_str[ 'title_Eng' ].isnull()].head(20)
```

# Na(결측값)을 포함함 모든 행에 대한 필터링

```python
# 각 행별로 결측값이 1개라도 있으면 True
movies_str.isnull().any()
```

```python
movies_str[ movies_str.isnull().any(axis=1) ].head(20)
```

# 특정 열의 값에 특정 문자열을 포함한 모습

* str.contains(): 특정문자열을 포함유무를 True/False로 반환

```python
movies_str[movies_str['title'].str.contains('코난')]
```

### 특정 단어 카운트

* count(): 문자열에 포함된 단어의 갯수를 반환

```python
movies_str['title'].str.count('아')
```
