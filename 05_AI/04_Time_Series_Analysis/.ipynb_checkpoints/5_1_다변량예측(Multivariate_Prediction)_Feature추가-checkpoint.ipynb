# 다변량 예측 (Multivariate Prediction)

* 주어진 시계열 데이터와 연관된 다른 특성을 활용하여 시계열 데이터 예측
* 여기서 활용하는 다른 특성을 시차 회귀변수(Lagged Regressor)라고 한다.
* 다변량 예측을 하기 위해서는 예측 값과 시차 회귀변수간 연관관계가 모델 성능향상에 명확한 관계가 있을 경우에 시도한다.  
  그렇지 않을 경우에는 일반적인 단일 변수 모델(Univaraiate Prediction)의 성능 보다 떨어진다.
* 시차 회귀변수는 NeuralProphet 모델인 경우 add_lagged_regressor 함수를 통해 추가한다.

```python
import pandas as pd

# Load the dataset for tutorial 4 with the extra temperature column
df = pd.read_csv("https://github.com/ourownstory/neuralprophet-data/raw/main/kaggle-energy/datasets/tutorial04.csv")
df.head()
```

```python
fig = df.plot(x="ds", y=["y", "temperature"], figsize=(10, 6))
```

추가 데이터를 확인한 후, 우리는 그것을 시차 회귀변수로 모델에 추가할 것입니다. 우리는 이전 튜토리얼에서의 모델을 시작으로 하고, 그 다음 온도의 시차 회귀변수를 추가하여 더 나은 에너지 가격 예측을 얻을 것입니다.

* plotly와 matplotlib 충돌 방지를 위해 코드 분리
