## 눈금 표시하기

### 기본사용

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

# 축설정
plt.xticks([0, 1, 2])
plt.yticks(np.arange(1, 6))

plt.show()
```

## 타이틀 설정하기

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

# axis: 설정할 축을 지정. 여기서 'both'는 x축과 y축 모두에 대해 설정하겠다는 것을 의미.
# direction: 틱의 방향을 지정. 'in'은 안쪽으로 틱이 향하도록 설정.
# length: 틱의 길이를 지정. 여기서는 3으로 설정되어 있으므로, 틱의 길이가 3으로 지정.
# pad: 눈금과 레이블 사이의 간격을 지정. 여기서는 6으로 설정되어 있으므로, 눈금과 레이블 사이의 간격이 6으로 지정.
# labelsize: 눈금 레이블의 크기를 지정. 여기서는 14로 설정되어 있으므로, 눈금 레이블의 크기가 14로 지정.
plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)

plt.title('Graph Title')

plt.show()
```

### 타이틀 위치와 오프셋 지정하기

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)

# 타이틀 위치와 오프셋 지정하기
plt.title('Graph Title', loc='right', pad=20)

plt.show()
```

* 여러개 타이틀 등록

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
plt.title('오른쪽 Title', loc='right', pad=20)

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
plt.title('왼쪽 Title', fontdict=title_font, loc='left', pad=20)

plt.show()
```

### 타이틀 정보 얻기

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color='#e35f62', marker='*', linewidth=2)
plt.plot(x, x**3, color='forestgreen', marker='^', markersize=9)

plt.tick_params(axis='both', direction='in', length=3, pad=6, labelsize=14)
title_right = plt.title('Left Graph Title', loc='right', pad=20)

title_font = {
    'fontsize': 16,
    'fontweight': 'bold'
}
title_left = plt.title('한글 타이틀', fontdict=title_font, loc='left', pad=20)

print(f'왼쪽 타이틀 위치: {title_left.get_position()}')
print(f'왼쪽 타이틀 이름: {title_left.get_text()}\n')

print(f'오른쪽 타이틀 위치: {title_right.get_position()}')
print(f'오른쪽 타이틀 위치: {title_right.get_text()}')

plt.show()
```

## 수평선/수직선 표시하기

### 수평선 그리기 - axhline([y위치, x시작점(퍼센트), x종료점(퍼센트)], 기타 스타일 옵션)

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

# plt.axhline(4.0, 0., 0.9, color='lightgray', linestyle='--', linewidth=2)
plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2)
# 여기서 0.1은 x축 너비를 좌측에서 10% 지점, 0.9는 x축 너비에서 좌측에서 90% 지점

plt.show()
```

* 마이너스 값 폰트 지원

```python
plt.rcParams['axes.unicode_minus'] = False  # 유니코드 마이너스 기호 사용
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.7, 0.95, color='lightgray', linestyle='--', linewidth=2)

plt.show()
```

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.5, 1, color='lightgray', linestyle='--', linewidth=2)

plt.show()
```

### 수평선 그리기 - hlines([y위치, x시작점, x종료점], 기타 스타일 옵션)
    * x 시작점, x 종료값: 절대값

```python
x = np.arange(0, 4, 0.5)

plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='-.', linewidth=2)
# plt.hlines(-0.62, 1.0, 2.5, color='gray', linestyle='solid', linewidth=3)
plt.hlines(-0.62, 1.5, 2.1, color='gray', linestyle='solid', linewidth=10)

plt.show()
```
