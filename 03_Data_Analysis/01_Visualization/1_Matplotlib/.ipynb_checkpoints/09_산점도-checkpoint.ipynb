### 레이블 추가

- x 좌표 (0.04)
텍스트를 표시할 x 좌표입니다. 여기서는 그래프의 x 축에서 0.04 (즉, 좌측에서 4%) 위치에 텍스트를 표시합니다. 이 값은 축의 비율로 표현됩니다.

- y 좌표 (0.96)
텍스트를 표시할 y 좌표입니다. 여기서는 그래프의 y 축에서 0.96 (즉, 위쪽에서 96%) 위치에 텍스트를 표시합니다. 이 값도 축의 비율로 표현됩니다.

- 텍스트 내용 (label_text)
그래프에 표시할 실제 텍스트입니다. 여기서는 품종과 색상 정보를 담고 있는 label_text 변수를 사용합니다.

- 글꼴 크기 (fontsize=12)
텍스트의 글꼴 크기를 설정합니다. 여기서는 폰트 크기를 12로 설정하였습니다.

- 수직 정렬 (verticalalignment='top')
텍스트의 수직 정렬 방식을 설정합니다. top으로 설정하면 지정된 y 좌표를 기준으로 텍스트 상단이 맞춰집니다.

- 수평 정렬 (horizontalalignment='left')
텍스트의 수평 정렬 방식을 설정합니다. left로 설정하면 지정된 x 좌표를 기준으로 텍스트 좌측이 맞춰집니다.

- 좌표 변환 (transform=plt.gca().transAxes)
좌표 변환을 현재 축의 비율로 설정합니다. plt.gca()는 현재의 축 객체를 반환하며, transAxes는 축의 비율 좌표계(0에서 1까지의 범위)를 의미합니다. 이를 통해 x와 y 좌표를 축의 비율로 지정할 수 있습니다. 즉, 그래프의 크기에 상관없이 텍스트의 위치가 일정하게 유지됩니다.

- 경계 상자 (bbox=dict(facecolor='white', alpha=0.5))

텍스트의 배경 상자를 설정합니다. bbox는 사전(dictionary) 형태로 설정하며, 여기서는 두 가지 속성을 사용했습니다:
facecolor='white': 배경 색상을 흰색으로 설정합니다.
alpha=0.5: 배경 색상의 투명도를 설정합니다. 0은 완전 투명, 1은 완전 불투명입니다. 여기서는 0.5로 설정하여 반투명 효과를 줍니다.

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd

# Load iris dataset from seaborn
iris = sns.load_dataset('iris')
iris.head()
iris['species']
color_map = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
# Define colors for each species
colors = iris['species'].map(color_map)
# Plot
plt.figure(figsize=(10, 6))
plt.scatter(iris['petal_length'], iris['petal_width'], c=colors)
plt.xlabel('꽃잎 길이')
plt.ylabel('꽃잎 너비')
plt.title('붓꽃 꽃잎 길이/너비 별 분포도')
label_text = "품종\n"

for species, color in color_map.items():
    label_text += f"{species}: {color}\n"
label_text = label_text[:-1]

plt.text(0.04, 0.96, label_text, fontsize=12, verticalalignment='top', 
         horizontalalignment='left', transform=plt.gca().transAxes, 
         bbox=dict(facecolor='white', alpha=0.5))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load iris dataset from seaborn
iris = sns.load_dataset('iris')

color_map = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
# Define colors for each species
colors = iris['species'].map(color_map)

# Plot
plt.figure(figsize=(10, 6))
plt.scatter(iris['petal_length'], iris['petal_width'], c=colors)
plt.xlabel('꽃잎 길이')
plt.ylabel('꽃잎 너비')
plt.title('붓꽃 꽃잎 길이/너비 별 분포도')

# Adding the text legend manually
plt.text(0.04, 0.96, "* 품종", fontsize=12, verticalalignment='top', 
         horizontalalignment='left', transform=plt.gca().transAxes, 
         )

plt.text(0.04, 0.92, "- setosa", fontsize=12, color='red', verticalalignment='top', 
         horizontalalignment='left', transform=plt.gca().transAxes)

plt.text(0.04, 0.88, "- versicolor", fontsize=12, color='green', verticalalignment='top', 
         horizontalalignment='left', transform=plt.gca().transAxes)

plt.text(0.04, 0.84, "- virginica", fontsize=12, color='blue', verticalalignment='top', 
         horizontalalignment='left', transform=plt.gca().transAxes)

plt.show()
```

* 범례 옵션을 활용한 데이터 재구성

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load iris dataset from seaborn
iris = sns.load_dataset('iris')

color_map = {'setosa': 'red', 'versicolor': 'green', 'virginica': 'blue'}
# Define colors for each species
colors = iris['species'].map(color_map)

# Plot
plt.figure(figsize=(10, 6))
for species, color in color_map.items():
    subset = iris[iris['species'] == species]
    plt.scatter(subset['petal_length'], subset['petal_width'], c=color, label=species, edgecolor='k')

plt.xlabel('꽃잎 길이')
plt.ylabel('꽃잎 너비')
plt.title('붓꽃 꽃잎 길이/너비 별 분포도')

# Adding the legend
plt.legend(title='품종', loc='upper left', edgecolor='black')

plt.show()
```

### 시나리오 설정

```python
np.random.seed(0)

n = 100
x = np.random.rand(n)
y = np.random.rand(n)
area = (50 * np.random.rand(n))**2

FIRST = 0
LAST = n-1
MEDIAN = int(n/2)
X_PADDING = 0.09
Y_PADDING = 0.07

plt.scatter(x, y, s=area, c='yellow')
plt.scatter(x[FIRST], y[FIRST], s=area[FIRST], c='red')
plt.text(x[FIRST]-X_PADDING, y[FIRST]+Y_PADDING, '1호 점포', fontdict={'size': 14})

plt.scatter(x[MEDIAN], y[MEDIAN], s=area[MEDIAN], c='green')
plt.text(x[MEDIAN]-X_PADDING, y[MEDIAN]+Y_PADDING, '중간 점포 ', fontdict={'size': 14})

plt.scatter(x[LAST], y[LAST], s=area[LAST], c='blue')
plt.text(x[LAST]-X_PADDING+0.06, y[LAST]+Y_PADDING+0.06, '가장 최신 점포', fontdict={'size': 14})
plt.title('더조은동 지역별, 상점별 매출 현황')
plt.xlabel('x 좌표 (동경 126 기준)')
plt.ylabel('y 좌표 (북위 37 기준)')
plt.show()
```

## 3차원 산점도

```python
n = 100
xmin, xmax, ymin, ymax, zmin, zmax = 0, 20, 0, 20, 0, 50
cmin, cmax = 0, 2

xs = np.array([(xmax - xmin) * np.random.random_sample() + xmin for i in range(n)])
ys = np.array([(ymax - ymin) * np.random.random_sample() + ymin for i in range(n)])
zs = np.array([(zmax - zmin) * np.random.random_sample() + zmin for i in range(n)])
color = np.array([(cmax - cmin) * np.random.random_sample() + cmin for i in range(n)])

fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c=color, marker='o', s=15, cmap='Greens')

plt.show()
```
