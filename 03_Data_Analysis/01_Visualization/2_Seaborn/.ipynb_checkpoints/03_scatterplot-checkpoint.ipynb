```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import font_manager,rc

import random

font_location = "C:\Windows\Fonts\malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
rc('font',family=font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
tips = sns.load_dataset("tips")
tips.head()
# size열은 일행을 의미함
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="time")
```

```python
tips.head()
```

```python
# style의 열이름을 지정하면 해당 열의 값을 표시하는 마커를 다르게 표시한다.
sns.scatterplot(data=tips, x="total_bill", y="tip", hue='time', style='time')
```

### 차트에서 얻은 인사이트

* 고가의 팁과 총 금액은 저녁 타임에서 발생된다.
* 총금액과 팁은 양의 상관관계를 보여주고 있다.

### 차트를 보고 얻은 인사이트

* 토요일 저녁시간에 고가의 팁이 발생한다.
* 토요일, 일요일에는 외식을 거의 저녁에 한다.
* 목요일 점심에 총금액 12 부근에 응집도가 높다.
* 목요일은 거의 점심식사를 한다.
* 금요일은 점심, 저녁식사를 모두 한다.
* 일요일 저녁에 적은 값에도 큰 팁을 내는 특이한 데이터가 존재한다.

### 인사이트에서 도출되는 비지니스 전략 

* 토요일 저녁시간에 고가의 팁이 발생한다.
  => 토요일에는 다채로운 고액 이벤트성 메뉴를 시도해 본다.
* 토요일, 일요일에는 외식을 거의 저녁에 한다.
  => 모든 고객이 저녁에 몰리기 때문에 추가적인 인력(아르바이트)에 대해 대비한다.
  => 주말 아침, 점심은 손님이 없어서 매장이 비는 상황이 발생한다. 이 시간에 매장 점검 또는 매장을 다른 목적(예: 대여)으로 활용하는 것에 대해 검토한다.
* 목요일은 거의 점심식사를 하고 목요일 점심에 총금액 12 부근에 응집도가 높다.
  => 저렴한 세트 메뉴를 더 개발한다. 
  => 추가 인력 배치도 고려한다.
* 금요일은 점심, 저녁시간 모두 방문하는 고객이 존재한다.
  => 점심, 저녁 메뉴 전환이 용이하도록 준비한다.
* 일요일 저녁에 적은 값에도 큰 팁을 내는 특이한 데이터가 존재한다.
  => 큰 팁을 내는 요인이 어떤 것이 있는지 추가 관찰 모니터가 필요

```python
# hue가 숫자 타입이라면 컬러맵을 자동으로 적용해 준다.
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size")
```

```python
# size속성에 크기를 표현 할 수 있는 열 이름을 지정한다.
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size", size='size')
```

* 붓꽃 예제

```python
# iris 데이터셋 불러오기
iris = sns.load_dataset('iris')

# 산점도 그리기
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

sns.scatterplot(x='petal_length', y='petal_width', hue='species', data=iris)
plt.title('Petal Length vs Petal Width')
plt.xlabel('Petal Length')
plt.ylabel('Petal Width')
plt.show()
```
