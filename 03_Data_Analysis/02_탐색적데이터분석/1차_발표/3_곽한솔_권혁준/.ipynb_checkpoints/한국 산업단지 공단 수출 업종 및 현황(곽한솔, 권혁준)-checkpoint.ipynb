# 4. 탐색적연구

#### step0] 가설, 데이터분석을 위한 목적 대상 또는 중가설을 세워본다.

* 무역 흑자는 계속 증가 중일것이다.  
* 무역 흑자가 가장 많이 나는 나라는 중국일것이다.  
* 무역 적자가 가장 많이 나는 나라는 EU일것이다.  
* 무역 흑자가 가장 많이 나는 항목은 상품수지일것이다.  
* 무역 적자가 가장 많이 나는 항목은 보험서비스수지일것이다.  
* 코로나시기에 여행수지는 줄었을 것이고 상품수지, 운송수지는 늘었을것이다.  
* 사드배치 관련시기에는 중국과의 경상수지가 줄어들었을것이다.  
* 중국의 가파른 성장으로 중국과의 무역수지는 줄어들고 있을것이다.

#### Step1] 우선순위가 높은 가설을 구체화한다.
* 예) 나이와 월급간에 상관관계가 있을 것이다. (40~60대 사이에서 급여가 가장 높을 것이다)

* 코로나 시기를 제외하면 경상수지가 지속적으로 증가하는 그래프를 그릴것이다.  
* 코로나 시기(2019~2022)는 경상수지가 줄어들었을것이다.  
* 사드배치 시기(2016~2017)는 중국과의 경상수지가 줄어들었을것이다.  
* 중국의 성장으로 중국과의 경상수지는 최근에 가파르게 줄어들었을것이다.

#### Step2] 가설증명: 데이터 재 구조화, 시각화를 통해 새로운 insight 및 정보를 습득 및 가설을 검증한다. 
* 예) 나이별 월급에 대한 평균값을 구하고 line차트로 시각화하여 추세를 파악하여 가설을 검증한다.
