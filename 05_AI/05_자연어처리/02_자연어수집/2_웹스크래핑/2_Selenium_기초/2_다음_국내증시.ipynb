```python
from bs4 import BeautifulSoup
import pandas as pd

# HTML 파일을 읽기 모드로 열기
with open("다음국내증시_24-08-22.html", "r", encoding="utf-8") as file:
    html_content = file.read()

# BeautifulSoup을 사용하여 HTML 파싱
soup = BeautifulSoup(html_content, "html.parser")
soup
```

```python
# 데이터 추출
data = []
# for item in soup.select("#boxDashboard"):
# for item in soup.select("#boxDashboard ul.tab.box_tabs li"): # CSS 선택자   li 태그를 중심으로 코스피, 코스닥, 코스피 200 항목이 반복된다.
for item in soup.select("#boxDashboard li"):
    title = item.find("h4").text.strip()
    price = item.find("strong").text.strip().replace(",", "")
    change = item.find_all("p")[0].text.strip()[1:]
    change_percent = item.find_all("p")[1].text.strip()[:-1]
    data.append([title, price, change, change_percent])

# 데이터 프레임 생성
df = pd.DataFrame(data, columns=["항목", "가격", "등락액", "등락률"])
df
```
