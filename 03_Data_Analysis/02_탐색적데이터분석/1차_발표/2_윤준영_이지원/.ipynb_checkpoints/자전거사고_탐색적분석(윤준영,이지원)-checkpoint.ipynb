```python
bike_accdent_pop['발생시간대'] = bike_accdent_pop['발생시간대'].str.replace('시','')
bike_accdent_pop['발생시간대'] = bike_accdent_pop['발생시간대'].astype('int64')
bike_accdent_pop['발생시간대']
```

```python
min_val = bike_accdent_pop['발생시간대'].min()
max_val = bike_accdent_pop['발생시간대'].max()
bins = np.linspace(min_val, max_val, 9) # 0~3시 3시간 간격
bike_accdent_pop['발생시간범주'] = pd.cut(bike_accdent_pop['발생시간대'], bins=bins, labels=[f"{bins[i]:.2f}~{bins[i+1]:.2f}" for i in range(len(bins)-1)], include_lowest=True)
bike_accdent_pop['발생시간범주'].head()
```

```python
bike_accdent_pop['발생시간범주'].info()
```

```python
grouped_df = bike_accdent_pop.groupby('발생시간범주')['부상자수'].sum().reset_index()
grouped_df
```

# * 가설1 : 부상자 수는 시간대에 영향을 받는다. 


```python
new_time_categories = ["0.00~3.00", "3.00~6.00", "6.00~9.00", "9.00~12.00", "12.00~15.00", "15.00~18.00", "18.00~21.00", "21.00~24.00"]
grouped_df['발생시간범주'] = new_time_categories[:len(grouped_df)]
grouped_df['발생시간범주']

grouped_df
```

```python
categories = grouped_df['발생시간범주']
values = [766, 832, 3667, 4813, 4833, 6195, 4826, 2322]

# 막대 그래프 생성
# fig: 그림, ax: 축 객체
fig, ax = plt.subplots()
bars = ax.bar(categories, values)


ax.bar_label(bars, fmt='%d', label_type='edge')

sns.barplot(data=grouped_df, x="발생시간범주", y="부상자수")
plt.xticks(rotation=45, ha='right',rotation_mode='anchor')


ax.set_title('발생시간별 부상자수')
ax.set_xlabel('발생시간범주')
ax.set_ylabel('부상자수')

plt.show()
```

* 특정 시간 자체보다는 그 시간대에 유동인구에 따라 부상자수가 많다.

# 가설 2 : 사고는 기상상태에 영향을 받는다

```python
def categorize_text(text):
    if '차대차' in text.lower():
        return '차대차'
    elif '차대사람' in text.lower():
        return '차대사람'
    elif '차량단독' in text.lower():
        return '차량단독'

```

```python
bike_accdent_pop['사고유형분류'] = bike_accdent_pop['사고유형'].apply(categorize_text)
bike_accdent_pop.head(50)
                     
```

```python
bike_wether = bike_accdent_pop[['사고내용', '사고유형분류','기상상태','가해자연령','피해자연령']]
bike_wether.head(5)
```

```python
bike_wether.loc[bike_wether['가해자연령'].isin(['불명', '없음']), '가해자연령'] = np.nan
bike_wether.loc[bike_wether['피해자연령'].isin(['불명', '없음']), '피해자연령'] = np.nan
```

```python
bike_wether.loc[:, '가해자연령'] = bike_wether['가해자연령'].astype(str).str.replace('세', '').astype(float)
bike_wether.loc[:, '피해자연령'] = bike_wether['피해자연령'].astype(str).str.replace('세', '').astype(float)
```

```python
bike_wether.head()
```

```python
sns.boxplot(data=bike_wether, x="가해자연령",y = '사고유형분류', hue='기상상태')
```
