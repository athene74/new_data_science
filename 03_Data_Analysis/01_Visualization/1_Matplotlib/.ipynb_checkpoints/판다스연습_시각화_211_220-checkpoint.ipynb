```python
import seaborn as sns
iris = sns.load_dataset("Iris")
iris.head()
```

```python
iris.groupby("species")["sepal_length"].mean().plot(kind='bar') # iris 데이터에서 species별 sepal_length의 평균을 구한 후 이를 'bar'타입으로 그리기
```

```python
iris.groupby("species").mean().plot.barh() # iris 데이터에서 품종별 특정값들의 평균을 구한 후 이를 barh 형태로 그리기
```

```python
iris.groupby('species').mean().T.plot.barh() # iris 데이터에서 품종별 특정값들의 평균을 구한 후 이를 transpose하고 barh 형태로 그리기
```

```python
s= iris.groupby('species')['sepal_length'].count()
s.name = ""
s.plot.pie(autopct="%.2f%%")
```

```python
import seaborn as sns

sns.scatterplot(
    x='petal_width',
    y='petal_length',
    hue='species',
    data=iris
)
```

```python
sns.boxplot(
    x='species',
    y='petal_length',
    data=iris
)
```

```python
sns.kdeplot(
    x='petal_length',
    hue='species',
    data=iris
)
```

```python
g=sns.FacetGrid(iris, col='species')
g.set(xlim=(0,10))
g.map(sns.histplot,'petal_length')
```
