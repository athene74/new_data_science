# 차트 구성 요소

* 산점도 (Scatter Plot):  
    * 각 데이터 포인트를 나타내는 점
    * 여기서 x 축은 bill_length_mm (부리 길이), y 축은 bill_depth_mm (부리 깊이)
* 추세선 (Regression Line):  
    * 데이터를 기반으로 한 선형 회귀선
    * 이 선은 bill_length_mm와 bill_depth_mm 간의 관계를 나타냄
* 추세선 부근의 연한 영역 (Confidence Interval):  
    * 추세선 주변의 연한 영역은 회귀선의 신뢰 구간(confidence interval)을 나타냄
    * 기본적으로, 이는 주어진 신뢰 수준(보통 95%)에서 회귀선이 실제로 있을 것으로 예상되는 범위를 의미
    * 이 영역은 회귀선의 불확실성을 시각적으로 표현

```python
sns.lmplot(data=penguins, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

```python
sns.lmplot(
    data=penguins, x="bill_length_mm", y="bill_depth_mm",
    hue="species", col="sex"  # 종별, 성별별 중분류까지는 나타내는 차트를 그릴 수 있다.
)
```
